```python
import pandas as pd

from sklearn.model_selection import train_test_split,GridSearchCV,KFold
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.feature_selection import SelectKBest,f_regression
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error
```

```python
data=pd.read_csv(r"G:\datasets\icicibank.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
data.drop(columns=['Date ','series '],inplace=True)
```

```python
data.isna().sum()
```

```python
for col in data.columns:
    data[col]=data[col].astype(str).str.replace(',','').astype(float).astype(int)
```

```python
data.info()
```

```python
X = data.drop(columns = ['close '])
y = data['close ']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#model1
```

```python
model1= Pipeline(steps=[
    ('standardscaler',StandardScaler()),
    ('linear',LinearRegression())
])
```

```python
model1.fit(X_train,y_train)
train_pred=model1.predict(X_train)
test_pred=model1.predict(X_test)
print("mean_absolute_error - train - model1 :",mean_absolute_error(y_train,train_pred))
print("mean_absolute_error - train - model1 :",mean_absolute_error(y_test,test_pred))
```

```python
#model2
```

```python
model2= Pipeline(steps=[
    ('standardscaler',StandardScaler()),
    ('poly',PolynomialFeatures(degree=2)),
    ('linear',LinearRegression())
])
```

```python
model2.fit(X_train,y_train)
train_pred=model2.predict(X_train)
test_pred=model2.predict(X_test)
print("mean_absolute_error - train - model2 :",mean_absolute_error(y_train,train_pred))
print("mean_absolute_error - train - model2 :",mean_absolute_error(y_test,test_pred))
```
